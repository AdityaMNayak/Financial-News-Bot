```python
def symbol_data(search_keyword,ticker): 
    res=Interval.in_daily
    exc='NSE'
    rolling_window_length=30
    req_data = pd.DataFrame(api.get_everything(qintitle= search_keyword,
                                               page_size = 100,
                                               sort_by='popularity'))

    news_df = pd.concat([req_data['articles'].apply(pd.Series)], axis = 1)
    final_news = news_df.loc[:, ['publishedAt', 'title','description']]
    final_news['publishedAt'] = pd.to_datetime(final_news['publishedAt'])
    final_news.sort_values(by = 'publishedAt', inplace = True, ascending=False)
    final_news.reset_index(inplace=True, drop=True)

    def lemmatizer(x):
        text = x.description

        text = text.replace("<ol><li>", "")

        text= text.lower()

        tokens = word_tokenize(text)
        tokens_without_stop = [word for word in tokens if word not in stopwords.words('english')]

        tokens_without_stop

        wordnet_lemmatizer = WordNetLemmatizer()
        lemmatized = [wordnet_lemmatizer.lemmatize(token) for token in tokens_without_stop]


        lemmatized_text = ' '.join(lemmatized)
        return lemmatized_text

    final_news['lemmatized_description'] = final_news.apply(lemmatizer, axis=1)


    scores = pd.DataFrame(final_news['lemmatized_description'].apply(vader.polarity_scores).tolist())
    final_news['compound'] = scores['compound'].values.tolist()
    final_news = final_news[final_news['compound'] != 0].reset_index(drop = True)

    def get_trade_open(date):
        curr_date_open = pd.to_datetime(date).floor('d').replace(hour=3, minute=45) - BDay(0)
        curr_date_close = pd.to_datetime(date).floor('d').replace(hour= 10,minute=0) - BDay(0)

        prev_date_close = (curr_date_open - BDay()).replace(hour=10,minute=0)
        next_date_open = (curr_date_close + BDay()).replace(hour=3,minute=45)

        if ((pd.to_datetime(date)>=prev_date_close) & (pd.to_datetime(date)<curr_date_open)):
            return curr_date_open
        elif ((pd.to_datetime(date)>=curr_date_close) & (pd.to_datetime(date)<next_date_open)):
            return next_date_open
        else:
            return None

    final_news['Trading_Time'] = final_news['publishedAt'].apply(get_trade_open)
    final_news.dropna(inplace = True)
    final_news['Date'] = pd.to_datetime(pd.to_datetime(final_news['Trading_Time']).dt.date)

    unique_dates = final_news['Date'].unique()
    grouped_dates = final_news.groupby(['Date'])
    keys_dates = list(grouped_dates.groups.keys())

    max_score = []
    min_score = []

    for key in grouped_dates.groups.keys():
        data_group = grouped_dates.get_group(key)
        if data_group["compound"].max() > 0:
            max_score.append(data_group["compound"].max())
        elif data_group["compound"].max() < 0:
            max_score.append(0)

        if data_group["compound"].min() < 0:
            min_score.append(data_group["compound"].min())
        elif data_group["compound"].min() > 0:
            min_score.append(0)

    extreme_score = pd.DataFrame({'Day':keys_dates, 'Min_Score':min_score, 'Max_Score':max_score})
    extreme_score['Final_Score'] = extreme_score[['Min_Score','Max_Score']].sum(axis = 1)
    #extreme_score.set_index('Day',drop=True,inplace=True)
    
    stock_data = tv.get_hist(ticker, 'NSE',interval= Interval.in_daily,n_bars=5000)
    stock_data.reset_index(inplace=True,drop=False)

    stock_data['Day'] = pd.to_datetime(pd.to_datetime(stock_data['datetime']).dt.date)

    stock_data['dailyReturns'] = (stock_data.close - stock_data.close.shift(1)) / stock_data.close.shift(1)

    returns_df = stock_data[['Day','symbol','dailyReturns']]

    merged_df = extreme_score[['Final_Score','Day']]
    merged_df = pd.merge(merged_df, returns_df, on='Day',how='left')


    
    return merged_df
```
